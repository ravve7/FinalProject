```python
import datetime
import random
import time
import names
import string
import radar 
import pandas as pd
```

```python
class Pizza:
    def __init__(self):
        self.price = 10 #set each pizza price as 10$
        self.topping = None

    def add_topping(self,toppings):
        self.topping = toppings
    def remove_topping(self):
        self.topping = None
    def pizza_crust(self):
        self.crust = random.choice(['thick','thin'])    
    def detail(self):
        return {'toppings':self.topping,
               'price':'{}$'.format(self.price),
               'crust':self.crust}
```

```python
class Customer:
    def customer_name(self,f_name,l_name):
        self.name = '{} {}'.format(f_name,l_name)
    def customer_phone(self,phone):
        self.phone = phone
    def customer_zipcode(self,zip_code):
        self.zip_code = zip_code
    def customer_freq_mil_num(self,frequent_milage_number):
        self.freq_mil_num = frequent_milage_number
```

```python
class PizzaOrder(Pizza,Customer):
    def __init__(self):
        Pizza.__init__(self)
        Customer.__init__(self)
        self.order = []

    def set_orderdate(self,date):
        self.order_date = date
    def add_pizza(self,name,*args):
        self.order.append(name)
        
    def remove_pizza(self,name):
        self.order.remove(name)
    
    def select_store(self):
        if self.zip_code == 95340:
            self.store = Store().address
        
    def promo_code(self, name):
        if name == 'CODE':
            self.promocode = 'applied'
            self.total_price = self.get_orderamount() - 30*self.get_orderamount()/100
            return self.total_price
        else:
            self.promocode = 'not applied'
            self.total_price = self.get_orderamount()
            return self.total_price

    def get_orderamount(self):
        self.total_price = len(self.order)*self.price
        return self.total_price
    def orderstatus(self):
        if self.total_price:
            print('ORDER_CREATED')
            print('ORDER_READY')
            print('ORDER_ON_DELIVERY')
            self.order_status = ('ORDER_COMPLETE')
            print("ORDER_COMPLETE")

        elif len(self.order) == 0:
            print('ORDER_CANCELED')
```

```python
class Store:
    def __init__(self):
        self.employees = []
        self.address = "papa tonys pizza, 15 street 95340"
        self.zip_code = 95340
        self.store_phone = 1234567890
    def add_emp(self, f_name, l_name):
        self.employees.append(Employee(f_name, l_name).name)
    def remove_emp(self, f_name, l_name):
        self.employees.remove(Employee(f_name, l_name).name)
    def store_detail(self):
        return {'address': self.address,
               'zip code': self.zip_code,
               'phone':self.phone}
```

```python
class Employee:
    def __init__(self, f_name, l_name):
        self.f_name = f_name
        self.l_name = l_name
        self.name = '{} {}'.format(self.f_name,self.l_name)
    def info(self):
        return "{} {}".format(self.f_name,self.l_name)      
```

### Generating Random orders with random data

```python
data = []
def order_robot():
    pizzas = ['pepporoni','Margherita','Carbonara','chicken_pizza','Hawaienne','4formages']
    toppings = ['Mushrooms', 'onions', 'Bacon','extra cheese']
    promo_code = ['CODE','code']

    for i in range(1,20):
        k = PizzaOrder()
        k.set_orderdate(radar.random_datetime(start = datetime.datetime(year=2020, month=1, day=1),
                                              stop = datetime.datetime.now()))
        k.customer_name(names.get_first_name(),names.get_last_name())
        k.customer_phone(''.join(random.choice(string.digits) for i in range(10)))
        k.customer_zipcode(95340)
        k.select_store()
        for j in range(random.randint(1,5)):
            k.add_pizza(random.choice(pizzas))
        k.add_topping(random.choice(toppings))
        k.pizza_crust()
        k.get_orderamount()
        k.promo_code(random.choice(promo_code))
        k.orderstatus()
        data.append({
            'order_date' : k.order_date,
            'customer_name' : k.name,
            "customer_phone" : k.phone,
            'customer_zipcode' : k.zip_code,
            'store' : k.store,
            'pizza' : k.order,
            'topping' : k.topping,
            'crust' : k.crust,
            'Promo Code' : k.promocode,
            'total_price($)' : k.total_price,
            'order_status' : k.order_status
        })
        print('order{} is completed suucessfully'.format(i))
```

```python
order_robot()
```
